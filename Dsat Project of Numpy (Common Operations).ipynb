```python
import pandas as pd 
import numpy as np
from skimage import io
```

```python
np_image=io.imread('mountain.jpg')
```

```python
print(type(np_image), '\t',np_image.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(np_image)                  #show
```

```python
np_image_flipped=np_image[:,::-1]       #Horizontal Flip
```

```python
plt.imshow(np_image_flipped)
```

```python
from PIL import Image
```

```python
np_image_flipped=np_image[::-1,:]                #Vertically
plt.imshow(np_image_flipped)
```

```python
from PIL import Image
import PIL.ImageOps
```

```python
np_image=Image.open('mountain.jpg')
a=PIL.ImageOps.invert(np_image)                #invert
```

```python
plt.imshow(a)
```

```python
from PIL import Image, ImageOps
     
im1 = Image.open("mountain.jpg")                   #grayscale
im2 = ImageOps.grayscale(im1)
plt.imshow(im2)
```

```python
from PIL import Image
img = Image.open('mountain.jpg')                            #black&white

thresh = 200
fn = lambda x : 255 if x > thresh else 0
r = img.convert('L').point(fn, mode='1')
plt.imshow(r)
```
